```python
import netCDF4 as nc
import os
import glob
import matplotlib.pyplot as plt
```

```python
sif_values_by_year = {}


for year in range(2007, 2017):
    year_str = str(year)  
    year_sif_values = []
    months = []
    folder_path = f"./SIF_data/GOME2A_SIF{year_str}"
   
    nc_files = sorted(glob.glob(os.path.join(folder_path, f"GOME2A_SIF{year_str}_*.nc")))
  
    for file in nc_files:
        filename = os.path.basename(file) 
        
        month_number = filename.split("_")[2][:2] 
        
        month_name = {
            "01": "January", "02": "February", "03": "March",
            "04": "April", "05": "May", "06": "June",
            "07": "July", "08": "August", "09": "September",
            "10": "October", "11": "November", "12": "December"
        }.get(month_number, "Unknown")

        dataset = nc.Dataset(file)
        
        sif_value = dataset.variables["SIF"][150, 224]

        year_sif_values.append(sif_value)
        months.append(month_name)
    
    # Store the year-wise SIF data in the dictionary
    sif_values_by_year[year_str] = (months, year_sif_values)

# Plotting all years' data
plt.figure(figsize=(15, 8))

# Loop over the years to plot
for year, (months, sif_values) in sif_values_by_year.items():
    plt.plot(months, sif_values, label=f"Year {year}", marker='o')

# Customizing the plot
plt.xlabel('Month')
plt.ylabel('SIF Value')
plt.title('SIF Values from 2007 to 2012')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better visibility
plt.legend(title="Years")
plt.tight_layout()  # Adjust layout to fit labels

# Show plot
plt.show()
```

```python
import os
import glob
#import h5py
import numpy as np
import netCDF4 as nc
import matplotlib.pyplot as plt

# Create a dictionary to store SIF values by year and pixel
sif_values_by_year_and_pixel = {}

# Loop through years 2007 to 2016
for year in range(2007, 2010):
    year_str = str(year)
    year_sif_values = []
    months = []
    folder_path = f"./SIF_data/GOME2A_SIF{year_str}"

    # Get all .nc files for the current year
    nc_files = sorted(glob.glob(os.path.join(folder_path, f"GOME2A_SIF{year_str}_*.nc")))

    for file in nc_files:
        filename = os.path.basename(file)  # Get filename

        # Extract the month from the filename
        month_number = filename.split("_")[2][:2]  # Get '01' from 'GOME2A_SIF2011_01.nc'
        
        # Convert month number to month name
        month_name = {
            "01": "January", "02": "February", "03": "March",
            "04": "April", "05": "May", "06": "June",
            "07": "July", "08": "August", "09": "September",
            "10": "October", "11": "November", "12": "December"
        }.get(month_number, "Unknown")

        # Open the NetCDF file
        dataset = nc.Dataset(file)

        # Extract the SIF values for all pixels (assuming it's a 2D grid of size (lat, lon))
        sif_values = dataset.variables["SIF"][:]  # Shape: (latitude, longitude)

        # Store the SIF values for the month
        year_sif_values.append(sif_values)  # Store the full grid for each month
        months.append(month_name)

    # Store the year-wise SIF data for all pixels in the dictionary
    sif_values_by_year_and_pixel[year_str] = {
        "months": months,
        "sif_values": year_sif_values  # Each month will have an entire grid of SIF values
    }

# Now sif_values_by_year_and_pixel contains the SIF values for each year and each month.

# Example: To plot the SIF value for a specific pixel (say, pixel at [100, 150]):
pixel_row, pixel_col = 150, 224

# Plotting SIF for this pixel across all years
plt.figure(figsize=(10, 6))

for year, data in sif_values_by_year_and_pixel.items():
    months = data["months"]
    sif_values = [month_sif[pixel_row, pixel_col] for month_sif in data["sif_values"]]
    plt.plot(months, sif_values, label=f"Year {year}")

# Customizing the plot
plt.xlabel('Month')
plt.ylabel(f'SIF Value (Pixel {pixel_row}, {pixel_col})')
plt.title('SIF Value Across Months for Pixel (100,150) from 2007 to 2016')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better visibility
plt.legend(title="Years")
plt.tight_layout()

# Show plot
plt.show()
```

```python
import numpy as np
import torch
from sklearn.preprocessing import StandardScaler

# Reshape the data into a matrix: rows = months, columns = years (SIF values)
months = list(sif_values_by_year["2007"][0])  # Month names
sif_values_matrix = []
# Create a matrix with each row representing a month's SIF values for all years
for month in months:
    month_sif_values = []
    for year in range(2007, 2017):
        if year == 2009:
            continue
        sif_values = sif_values_by_year[str(year)][1]
        month_index = sif_values_by_year[str(year)][0].index(month)
        month_sif_values.append(sif_values[month_index])
    sif_values_matrix.append(month_sif_values)

sif_values_matrix = np.array(sif_values_matrix).T
#Normalize the data?
```

```python
import torch.nn as nn
import torch.optim as optim

class Autoencoder(nn.Module):
    def __init__(self, input_dim, hidden_dim = 3):
        super(Autoencoder, self).__init__()
        # Encoder: compress the input data
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 5),
            nn.ReLU(),
            nn.Linear(5, hidden_dim),
            nn.ReLU()
        )
        # Decoder: reconstruct the input data
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dim, input_dim),
            nn.Sigmoid()  # Use sigmoid for reconstruction to keep the output between 0 and 1
        )
    
    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
# Convert data to PyTorch tensors
X_train = torch.tensor(sif_values_matrix, dtype=torch.float32)

# Initialize the model, loss function, and optimizer
input_dim = X_train.shape[1]  # Number of years (features)
hidden_dim = 3  # You can try different sizes for the hidden layer
model = Autoencoder(input_dim, hidden_dim)

# Loss function and optimizer
criterion = nn.MSELoss()  # Mean Squared Error loss for reconstruction error
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
epochs = 10000
for epoch in range(epochs):
    model.train()
    
    # Zero the gradients
    optimizer.zero_grad()
    
    # Forward pass
    outputs = model(X_train)
    
    # Calculate the reconstruction loss
    loss = criterion(outputs, X_train)
    
    # Backward pass and optimization
    loss.backward()
    optimizer.step()
    
    if epoch % 1000 == 0:
        print(f"Epoch [{epoch}/{epochs}], Loss: {loss.item():.4f}")
```

```python
all_years_sif = []

for year in range(2007, 2017):
    for year in range(2007, 2017):
        if year == 2009:
            continue
    if str(year) in sif_values_by_year:
        all_years_sif.append(sif_values_by_year[str(year)][1])  # Extract SIF values

# Convert to NumPy array for easy averaging
all_years_sif = np.array(all_years_sif)  # Shape: (number_of_years, 12)

# Compute the mean SIF value for each month across all years
expected_sif_pattern = np.mean(all_years_sif, axis=0)  # Shape: (12,)

# Plot the expected SIF pattern
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(range(1,13), expected_sif_pattern, marker='o', linestyle='-', color='blue', label="Expected Yearly SIF")
plt.xticks(range(1,13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.xlabel("Month")
plt.ylabel("SIF Value")
plt.title("Expected SIF Yearly Pattern")
plt.legend()
plt.show()
```

```python
# Convert the expected SIF pattern to a tensor
sif_tensor = torch.tensor(expected_sif_pattern, dtype=torch.float32).unsqueeze(0)  # Shape: (1, 12)

# Get the reconstructed "ideal" yearly SIF shape
with torch.no_grad():
    reconstructed_sif = model(sif_tensor).squeeze().numpy()

# Plot the expected vs reconstructed SIF pattern
plt.figure(figsize=(8, 5))
plt.plot(expected_sif_pattern, marker='o', linestyle='-', color='blue', label="Original Expected SIF")
plt.plot(reconstructed_sif, marker='o', linestyle='--', color='green', label="Reconstructed SIF (Autoencoder)")

# Customizing plot
plt.xticks(range(1, 13), ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.xlabel("Month")
plt.ylabel("SIF Value")
plt.title("Most Likely Yearly SIF Shape (Autoencoder)")
plt.legend()
plt.show()
```

```python
# Function to calculate reconstruction error
def calculate_reconstruction_error(autoencoder, data):
    
    # Convert data to tensor
    data_tensor = torch.tensor(data, dtype=torch.float32)
    # Get the reconstruction from the autoencoder
    reconstructed_data = autoencoder(data_tensor)
    
    # Calculate the Mean Squared Error (MSE) between the original data and reconstructed data
    mse = nn.MSELoss()
    error = mse(reconstructed_data, data_tensor).item()  # Convert to a scalar value
    return error

calculate_reconstruction_error(Autoencoder(12,3), np.array(sif_values_by_year["2009"][1]))
```

```python
import torch
import torch.nn as nn

# Function to calculate reconstruction error for each month
def calculate_monthly_reconstruction_error(autoencoder, year_sif_values):
    # Initialize a list to store the reconstruction error for each month
    monthly_errors = []
    
    for month_sif_values in year_sif_values:
        # Convert the SIF values of the month to a tensor
        month_sif_tensor = torch.tensor(month_sif_values, dtype=torch.float32).unsqueeze(0)  # Add batch dimension
        
        # Get the reconstruction for the month
        reconstructed_month_sif = autoencoder(month_sif_tensor)
        
        # Calculate the Mean Squared Error (MSE) for this month
        mse = nn.MSELoss()
        error = mse(reconstructed_month_sif, month_sif_tensor).item()
        
        # Append the error to the list
        monthly_errors.append(error)
    
    return monthly_errors

# Example usage for a particular year, say 2009
year = 2009
sif_values = np.array(sif_values_by_year[str(year)][1]) # Get the SIF values for the year (you already have this)

# Call the function to get monthly reconstruction errors
monthly_errors_2009 = calculate_monthly_reconstruction_error(Autoencoder(12,3), sif_values)

# Print the errors for each month
print(f"Monthly reconstruction errors for {year}:")
for month, error in zip(sif_values_by_year[str(year)][0], monthly_errors_2009):
    print(f"{month}: {error}")
```
